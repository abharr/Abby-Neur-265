```python
#import modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
```

```python
#import data
fear_traces = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/fear_conditioning_traces_hw8.csv',delimiter=',')
extinction_predictors = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/extinction_predictors_hw8.csv',delimiter=',')
shock_index = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/shock_index.txt', delimiter = ',').astype(int)
tone_index = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/tone_index.txt',delimiter=',').astype(int)
```

```python
#separate fear_traces into conditioning and extinction
conditioning_traces = fear_traces[0:10458]
extinction_traces = fear_traces[10458:]
```

```python
#print size of conditioning_traces and extinction_traces
print(conditioning_traces.shape)
print(extinction_traces.shape)
```

```python
#create time variables
time1 = 10458/15
time2 = 14210/15
time_conditioning = np.linspace(0,time1,num=10458)
time_extinction = np.linspace(0,time2,num=14210)
print(time1)
print(time2)
print(time1/60)
print(time2/60)
```

63 ROIs were imaged in this recording session.

The conditioning recording session lasted 697 seconds, or 11.6 minutes. 

The extinction recording session lasted 947 seconds, or 15.8 minutes.

```python
#plot first 2 ROIs from conditioning and extinction
fig, ax = plt.subplots(2)
ax[0].plot(time_conditioning,conditioning_traces[:,0],color='slateblue')
ax[0].plot(time_conditioning,conditioning_traces[:,1],color='maroon')
ax[0].plot([shock_index[0],shock_index[0]],[-0.1,1],color='black')
ax[0].plot([shock_index[1],shock_index[1]],[-0.1,1],color='black')
ax[0].plot([shock_index[2],shock_index[2]],[-0.1,1],color='black')
ax[0].plot([tone_index[0],tone_index[0]],[-0.1,1],color='forestgreen')
ax[0].plot([tone_index[1],tone_index[1]],[-0.1,1],color='forestgreen')
ax[0].plot([tone_index[2],tone_index[2]],[-0.1,1],color='forestgreen')
ax[1].plot(time_extinction,extinction_traces[:,0],color='slateblue')
ax[1].plot(time_extinction,extinction_traces[:,1],color='maroon')
```

Both ROIs responded similarly during conditioning extinction, with the second ROI showing slightly more fluorescence at some point. During conditioning, both ROIs responded with higher activity right after a shock, although the response decreased with each successive shock.

```python
#modeling: create shock and tone array and predictors variables
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0
     
predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array,'shock onset': shock_array})
```

```python
#model conditioning data
r_model_conditioning = np.zeros(63)

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model_conditioning = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model_conditioning_results = model_conditioning.fit()
  predicted_trace = model_conditioning_results.predict()
  r_conditioning = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r_conditioning.statistic
```

```python
#model extinction data
r_model_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  model_extinction = sm.GLM(trace_temp, extinction_predictors[1:], family=Poisson())
  model_extinction_results = model_extinction.fit()
  predicted_trace = model_extinction_results.predict()
  r_extinction = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r_extinction.statistic
```

```python
#correlate extinction
correlation_stat, correlation_p = pearsonr(r_model_conditioning,r_model_extinction)
print('Pearson r value is',correlation_stat)
```

```python
#scatter plot with conditioning and extinction
plt.scatter(r_model_conditioning,r_model_extinction)
```

There does not seem to be any kind of strong correlation between cell response to shock and tone onset during conditioning and cell response to tone during extinction. The Pearson r value is very low (0.01), and the scatter plot does not show any pattern or linearity.
