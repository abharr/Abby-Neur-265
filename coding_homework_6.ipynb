```python
#import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import data
acc_hits = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/acc_hits_hw6%20(1).csv',delimiter=',')
acc_fa = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/acc_fa_hw6%20(1).csv',delimiter=',')
lc_hits = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/lc_hits_hw6%20(1).csv',delimiter=',')
lc_fa = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/lc_fa_hw6%20(1).csv',delimiter=',')
```

```python
#print shape of each array
print(acc_hits.shape)
print(acc_fa.shape)
print(lc_hits.shape)
print(lc_fa.shape)
```

```python
#create time variable
time = np.linspace(0,4,num=8001)
```

```python
#find sampling rate
r,c = acc_hits.shape
srate = c/4
print('Sampling rate is',srate,'Hz')
```

```python
#try acc and lc data in single subplot
fig1, ax1 = plt.subplots(1,2,figsize=(12,4))
ax1[0].plot(time, acc_hits[0],color='firebrick',label='ACC')
ax1[0].plot(time, lc_hits[0],color='darkslateblue',label='LC')
ax1[0].legend()
ax1[0].set(ylabel='voltage (uV)',xlabel='time (s)',title='Hit Condition')
ax1[1].plot(time, acc_fa[0],color='firebrick',label='ACC')
ax1[1].plot(time, lc_fa[0],color='darkslateblue',label='FC')
ax1[1].set(xlabel='time (s)',ylabel='voltage (uV)',title='False Alarm Condition')
ax1[1].legend()
```

The dominant oscillation in the ACC appears to be around 3-5 Hz. The dominant oscillation in the LC is difficult to pick out, but appears to be somewhere around 7 Hz. While the LC does not exhibit much syncronous activity in either condition, there is a clear difference between the ACC LFP recordings for the hit and false alarm conditions. For the hit condition, the waves are large and mostly regular throughout the recording. For the false alarm, the waves begin small and more random, and then become larger around t=1.6

```python
#compute coherence for hit condition

dt = time[1] - time[0] 
K = acc_hits.shape[0]              
N = acc_hits.shape[1]              

xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in acc_hits
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in lc_hits


T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0) 
Syy = scale * (yf * yf.conj()).mean(0)  
Sxy = scale * (xf * yf.conj()).mean(0)  

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                    
plt.plot(f, cohr.real)                   
plt.xlim([0, 20])                           
plt.ylim([0, 1])                            
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

There is the most phase coherence around 1 and 5-7 Hz.

```python
#compute coherence for false alarm condition

dt = time[1] - time[0] 
K1 = acc_fa.shape[0]              
N1 = acc_fa.shape[1]              

xf1 = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])
yf1 = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])

 
T = time[-1]
scale = 2 * dt**2 / T
Sxx1 = scale * (xf1 * xf1.conj()).mean(0)  
Syy1 = scale * (yf1 * yf1.conj()).mean(0) 
Sxy1 = scale * (xf1 * yf1.conj()).mean(0)

cohr = np.abs(Sxy1) / (np.sqrt(Sxx1) * np.sqrt(Syy1))

f = np.fft.rfftfreq(N1, dt)                    
plt.plot(f, cohr.real)                
plt.xlim([0, 20])                         
plt.ylim([0, 1])                     
plt.xlabel('Frequency [Hz]')          
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

In the false alarm recordings, there is the most phase coherence at ~1 Hz.

There is more overall coherence during the hit condition, which means that the ACC and LC "talk" to each other more during a hit than a false alarm.
