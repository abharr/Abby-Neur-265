```python
#import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load data files
url1 = 'https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/acc_hits_hw5.csv'
url2 = 'https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/acc_fa_hw5.csv'
acc_hits = np.loadtxt(url1, delimiter=',')
acc_fa = np.loadtxt(url2, delimiter=',')
```

```python
#print shape of each array
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#create time variable and print its shape
time = np.linspace(0,4,num=8001)
print(time.shape)
```

```python
#determine sampling rate

s_rate = 8001/4
print('sampling rate is',s_rate,'Hz')
```

```python
#create subplots of the first trial for acc_hits and acc_fa
fig, ax = plt.subplots(1,2,)
ax[0].plot(time,acc_hits[0],color='darkblue')
ax[0].set_ylabel('Voltage (uV)')
ax[0].set_xlabel('Time(s)')
ax[1].plot(time,acc_fa[0],color='maroon')
ax[1].set_xlabel('Time(s)')
```

The dominant oscillation appears to be around 3-4 Hz for the hit condition and 1 Hz for the false alarm condition. For acc_hits the amplitude increases dramatically at about 2 sec, while the amplitude for acc_fa stays smaller and more constant.

```python
#create ERP for acc_hits
plt.plot(time, acc_hits.mean(0))
plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')
plt.title('ERP of hit condition')
plt.plot([2,2],[-40,40],'k',lw=2)
plt.plot([0,4],[0,0],'k',lw=2)
```

The amplitude of the LFP oscillations appears to increase dramatically after the presentation of the S+ stimulus under the hit condition.

```python
#create ERP for acc_fa
plt.plot(time, acc_fa.mean(0),color='darkred')
plt.plot(time, acc_hits.mean(0),color='darkblue')
plt.xlabel('Time [s]')          
plt.ylabel('Voltage [$\mu V$]')
plt.title('ERP of S+ and S- conditions')
plt.plot([2,2],[-40,40],'k',lw=2)
plt.plot([0,4],[0,0],'k',lw=2)
```

Unlike in the hit condition, there does not seem to be a change in the oscillations following the presentation of the S- for the false alarm condition.
