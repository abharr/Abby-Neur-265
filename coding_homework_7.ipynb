```python
#import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import data
spikes_right_trial1 = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/spikes_right_trial1.csv',delimiter=',')
spikes_left_trial1 = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/spikes_left_trial1.csv',delimiter=',')
spikes_right_trial2 = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/spikes_right_trial2.csv',delimiter=',')
spikes_left_trial2 = np.loadtxt('https://raw.githubusercontent.com/abharr/Abby-Neur-265/main/spikes_left_trial2.csv',delimiter=',')
```

```python
#print length of each array
print('Spikes_right_trial1 length is:',len(spikes_right_trial1))
print('Spikes_left_trial1 length is:',len(spikes_left_trial1))
print('Spikes_right_trial2 length is:',len(spikes_right_trial2))
print('Spikes_left_trial2 length is:',len(spikes_left_trial2))
```

```python
#determine firing rate of neuron during each trial
fr_right_1 = len(spikes_right_trial1)/30
fr_right_2 = len(spikes_right_trial2)/30
fr_left_1 = len(spikes_left_trial1)/30
fr_left_2 = len(spikes_left_trial2)/30
print('Firing rate during right turn trial 1:',fr_right_1,'Hz')
print('Firing rate during right turn trial 2:',fr_right_2,'Hz')
print('Firing rate during left turn trial 1:',fr_left_1,'Hz')
print('Firing rate during left turn trial 2:',fr_left_2,'Hz')
```

There does not seem to be a difference in firing rate of this neuron before a right turn vs a left turn.

```python
#make spike trails of the two trials
fig, ax = plt.subplots(2)
fig.tight_layout(h_pad=4)
ax[0].plot(spikes_right_trial1,2*np.ones_like(spikes_right_trial1),'.',color='slateblue',label='Right Turn')
ax[0].plot(spikes_left_trial1,np.ones_like(spikes_left_trial1),'.',color='maroon',label='Left Turn')
ax[0].set_xlabel('time(s)')
ax[0].set_ylim(0,3)
ax[0].set_title('Trial 1')
ax[1].plot(spikes_right_trial2,2*np.ones_like(spikes_right_trial2),'.',color='slateblue',label='Right Turn')
ax[1].plot(spikes_left_trial2,np.ones_like(spikes_left_trial2),'.',color='maroon',label='Left Turn')
ax[1].set_xlabel('time(s)')
ax[1].set_ylim(0,3)
ax[1].set_title('Trial 2')
ax[0].legend()
ax[1].legend()
ax[0].set_yticks([],[])
ax[1].set_yticks([],[])
```

Each trial looks slightly different in terms of where the spikes cluster, but overall the neuron looks slightly more "bursty" during left turns. It is hard to see a clearcut difference however.

```python
#calculate ISI arrays
ISI_right1 = np.diff(spikes_right_trial1)
ISI_left1 = np.diff(spikes_left_trial1)
ISI_right2 = np.diff(spikes_right_trial2)
ISI_left2 = np.diff(spikes_left_trial2)
```

```python
#plot histograms of ISI values for each trial
fig1, ax1 = plt.subplots(2,2)
bins = np.arange(0,2,0.01)
ax1[0,0].hist(ISI_left1,bins,label='Trial 1: Left Turn',color='salmon')
ax1[0,1].hist(ISI_right1,bins,label='Trial 1: Right Turn',color='purple')
ax1[1,0].hist(ISI_left2,bins,label='Trial 2: Left Turn',color='darkred')
ax1[1,1].hist(ISI_right2,bins,label='Trial 2: Right Turn',color='darkslateblue')
ax1[0,0].legend()
ax1[0,1].legend()
ax1[1,0].legend()
ax1[1,1].legend()
```

All the histograms look about the same. For both left and right turns, the ISIs tend to cluster around 0.1 seconds.

```python
#calculate firing rate during trials
binsize = 1
bins2 = np.arange(0,30,1)
fr_r1 = np.zeros(len(bins2)-1)                                             # Initialize our firing rate array                                 
for i in range(len(bins2)-1):                                               # Loop through each bin
  index = (spikes_right_trial1 > bins2[i]) & (spikes_right_trial1 < bins2[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial1[index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_r1[i] = (len(true_values))/1
fr_l1 = np.zeros(len(bins2)-1)  
for i in range(len(bins2)-1):                                               # Loop through each bin
  index = (spikes_left_trial1 > bins2[i]) & (spikes_left_trial1 < bins2[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial1[index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_l1[i] = (len(true_values))/1
fr_r2 = np.zeros(len(bins2)-1)   
for i in range(len(bins2)-1):                                               # Loop through each bin
  index = (spikes_right_trial2 > bins2[i]) & (spikes_right_trial2 < bins2[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial2[index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_r2[i] = (len(true_values))/1
fr_l2 = np.zeros(len(bins2)-1)  
for i in range(len(bins2)-1):                                               # Loop through each bin
  index = (spikes_left_trial2 > bins2[i]) & (spikes_left_trial2 < bins2[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial2[index]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_l2[i] = (len(true_values))/1
```

```python
#plot firing rates across time
fig2, ax2 = plt.subplots(2)
ax2[0].plot(bins2[:-1],fr_r1,label='Trial 1 Right Turn',color='slateblue')
ax2[0].plot(bins2[:-1],fr_l1,label='Trial 1 Left Turn',color='maroon')
ax2[1].plot(bins2[:-1],fr_r2,label='Trial 2 Right Turn',color='slateblue')
ax2[1].plot(bins2[:-1],fr_l2,label='Trial 2 Left Turn',color='maroon')
ax2[0].legend()
ax2[1].legend()
```

The firing rate of the neuron over time varies between trials and between left vs right turns. The trend that remains consisten between trials is that the firing rate before a right turn is highest in the first 15 seconds. Otherwise, there are no easily observable trends during right vs left turns.
